# Darren Seals

This notebook explores Darren Seals ([@kingdseals](https://twitter.com/kingdseals)) related tweets that were collected with [twint](https://github.com/twintproject/twint).

## Read in Data

First we need to read in the data that we collected as CSV.

```python
import pandas

df = pandas.read_csv('data/kingdseals.csv', parse_dates=['date'], index_col=['date'], low_memory=False)
len(df)
```

```python
df.head(5)
```

## Tweets by Darren Seals

Now we can look at tweets that he sent over time.

```python
his = df[df['username'] == 'kingdseals']
his = his['id'].resample('W').count()
his.name = 'tweets'
his
```

And we can graph it.

```python
import altair

chart = altair.Chart(his.reset_index(), 
    width=800, 
    title='KingDSeals Tweets per Week')
chart = chart.mark_line().encode(x='date:T', y='tweets:Q')
chart

```

## Tweets About KingDSeals

We can also look at tweets mentioning @KingDSeals.

```python
others = df[df['username'] != 'kingdseals']
others = others.resample('W')['id'].count()
chart = altair.Chart(others.reset_index(), width=800, title='KingDSeals Mentions')
chart = chart.mark_line().encode(x='yearmonth(date):T', y='id:Q')
chart
```

## Mentions After He Died

We can look at what users were tweeting about him most in the week after he died on September 6, 2016.

```python
died = df.query('date >= "2016-09-06" and date <= "2016-09-13"')
died = died.resample('1D').count()
chart = altair.Chart(died.reset_index(), width=800, title='')
chart = chart.mark_line().encode(x='day(date):T', y='id:Q')
chart
```

## Top Mentioning Users

We can also see what users mentioned him the most in that week.

```python
mentions = df.query('date >= "2016-09-06" and date <= "2016-09-13"')
mentions = mentions.groupby('username').count()
mentions = mentions.sort_values('id', ascending=False)['id']
mentions.name = 'tweets'

for username, count in mentions.items():
    print('https://twitter.comn/{0:s} {1:n}'.format(username, count))
    if count == 1:
        break
```

## Most Retweeted

What were Darren Seals most retweeted tweets?

```python
his = df.reset_index().query('username == "kingdseals"')
his = his.sort_values('retweets_count', ascending=False)
for id, row in his.iterrows():
    print('https://twitter.com/kingdseals/status/{} {}'.format(row['id'], row['retweets_count']))
    if row['retweets_count'] < 5:
        break
```

## Most Replied To

What tweets of Darren Seals were most replied to?

```python
his = df.reset_index().query('username == "kingdseals"')
his = his.sort_values('replies_count', ascending=False)
for id, row in his.iterrows():
    print('https://twitter.com/kingdseals/status/{} {}'.format(row['id'], row['replies_count']))
```

## Who Did Seals Respond to the Most?

This is a bit tricky because there can be more than one value stored in df['mentions'], as a list:

    ['abc', '123']

```python

import json
from collections import Counter

his = df.reset_index().query('username == "kingdseals"')
counter = Counter()
for i, row in his.iterrows():
    for user in json.loads(row['mentions'].replace("'", '"')):
        counter[user] += 1
        
for user, count in counter.most_common(25):
    print('https://twitter.com/{}'.format(user), count)
```
